```python
from selenium import webdriver
from selenium.webdriver import ChromeOptions
import time
```

```python
options=ChromeOptions()
options.add_argument('--headless=new')
```

```python
url="https://www.amazon.in/Sony-CFI-2008A01X-PlayStation%C2%AE5-Console-slim/dp/B0CY5HVDS2/ref=sr_1_1?sr=8-1"
```

```python
with webdriver.Chrome(options=options) as driver:
    driver.get(url) # HTTP GET Request -> request.get(url)
    time.sleep(10)
    html=driver.page_source
```

```python
print(html)
```
